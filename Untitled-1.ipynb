```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import yfinance as yf
from arch import arch_model

ticker = 'AMD'
start_date = '2013-01-01'
end_date = '2023-01-01'

# Load the CSV files
stock_data = yf.download(tickers=ticker, start=start_date, end=end_date)
sp500_data = yf.download(tickers='SPY', start=start_date, end=end_date)
```

```python
# Convert 'Date' to datetime and set as index
stock_data.index = pd.to_datetime(stock_data.index)
sp500_data.index = pd.to_datetime(sp500_data.index)

# Rename 'Adj Close' columns for clarity
stock_data.rename(columns={'Adj Close': f'{ticker} Adj Close'}, inplace=True)
sp500_data.rename(columns={'Adj Close': 'SP500 Adj Close'}, inplace=True)

# Merge the two datasets on the 'Date' column using an outer join
combined_data = pd.merge(stock_data[[f'{ticker} Adj Close']], sp500_data[['SP500 Adj Close']], 
                         left_index=True, right_index=True, how='inner')

# Calculate daily returns for each dataset
combined_data[f'{ticker} Returns'] = combined_data[f'{ticker} Adj Close'].pct_change()
combined_data['SP500 Returns'] = combined_data['SP500 Adj Close'].pct_change()

# Calculate alpha (AAPL return - SP500 return)
combined_data['Alpha'] = combined_data[f'{ticker} Returns'] - combined_data['SP500 Returns']

# Fit a GARCH(1,1) model to the alpha returns
garch_model = arch_model(combined_data['Alpha'].dropna() * 100, vol='Garch', p=1, q=1)
model_result = garch_model.fit(disp='off')

# Predict the conditional volatility
combined_data['Volatility'] = model_result.conditional_volatility / 100

# Calculate the positive and negative volatility envelopes
combined_data['Positive Envelope'] = combined_data['Alpha'].mean() + combined_data['Volatility']
combined_data['Negative Envelope'] = combined_data['Alpha'].mean() - combined_data['Volatility']

# Plotting the alpha with volatility envelopes
plt.figure(figsize=(10, 6))
plt.plot(combined_data.index, combined_data['Alpha'], label='Alpha')
plt.plot(combined_data.index, combined_data['Positive Envelope'], label='Volatility', color='red')
plt.plot(combined_data.index, combined_data['Negative Envelope'], color='red')
plt.title(f'{ticker} Alpha Over S&P 500')
plt.xlabel('Date')
plt.ylabel('Returns')
plt.legend()
plt.grid(True)
plt.show()

# Simulate capital growth based on alpha
initial_capital = 100  # Assuming an initial investment of $1
combined_data['Capital'] = initial_capital * (1 + combined_data['Alpha']).cumprod()
combined_data['Slow MA'] = combined_data['Capital'].rolling(10).mean()
combined_data['Fast MA'] = combined_data['Capital'].rolling(50).mean()

# Plotting the capital curve
plt.figure(figsize=(10, 6))
plt.yscale("log")
plt.plot(combined_data['Capital'], color='black', label='Price')
plt.plot(combined_data['Slow MA'], color='blue', label='Slow MA')
plt.plot(combined_data['Fast MA'], color='orange', label='Fast MA')
plt.title('Capital Growth Curve based on Alpha')
plt.xlabel('Date')
plt.ylabel('Capital')
plt.legend()
plt.grid(True)
plt.show()
```
